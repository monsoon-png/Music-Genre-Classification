Here, we plot the waveform of one example song each of every genre, to understand how they look

```python
import os
import librosa
import matplotlib.pyplot as plt
```

```python
def waveform_plot(input_path, genre,output_dir, sr=22050):
    y, sr = librosa.load(input_path, sr=sr)
    plt.figure(figsize=(16, 5))
    librosa.display.waveshow(y, sr=sr,color = "#6600CC")
    plt.title(f'Waveform of a {genre.capitalize()} Song')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    output_path = os.path.join(output_dir, f"{genre}_waveform.png")
    plt.savefig(output_path)
    plt.show()
    plt.close()  
    print(f"Saved {genre} waveform to {output_path}")

```

```python
input_path = "/Users/aravpatel/iml final/genres"
output_dir = "wavplots"
os.makedirs(output_dir, exist_ok=True) 
genres = [folder for folder in os.listdir(input_path) if not folder.startswith(".")]
```

```python
for genre in genres:
    genre_path = os.path.join(input_path, genre)
    audio_files = [f for f in os.listdir(genre_path) if f.endswith('.wav')]
    if audio_files:  # Ensure there are files in the genre folder
        file_path = os.path.join(genre_path, audio_files[0])  # Take the first file
        waveform_plot(file_path, genre, output_dir)
```
