This Notebook plots the other features extracted using an example

```python
import librosa
import os
import librosa.display
import matplotlib.pyplot as plt
```

```python
file_path = "/Users/aravpatel/iml final/genres/hiphop/hiphop.00001.wav"
y, sr = librosa.load(file_path, sr=22050)
output_folder = "other_features"
os.makedirs(output_folder, exist_ok=True)  
```

```python
output_file = os.path.join(output_folder, "zcr_plot.png")
zcr = librosa.feature.zero_crossing_rate(y)
plt.figure(figsize=(15, 6))
plt.plot(zcr[0], label='Zero Crossing Rate')
plt.title('Zero Crossing Rate (ZCR)')
plt.xlabel('Frames')
plt.ylabel('Rate')
plt.tight_layout()
plt.savefig(output_file)
plt.show()
plt.close()  
```

```python
rms = librosa.feature.rms(y=y)
output_file = os.path.join(output_folder, "rms_plot.png")
plt.figure(figsize=(15, 6))
plt.plot(rms[0], label='RMS Energy',color='red')
plt.title('Root Mean Square (RMS) Energy')
plt.xlabel('Frames')
plt.ylabel('Energy')
plt.grid()
plt.tight_layout()
plt.savefig(output_file)
plt.show()
plt.close()  
```

```python
def plot_chromagram(file_path, output_folder, sr=22050):
    # Load the audio file
    y, sr = librosa.load(file_path, sr=sr)
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    output_file = os.path.join(output_folder, "chroma_plot.png")
    plt.figure(figsize=(12, 8))
    librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', sr=sr, cmap='magma')
    plt.colorbar()
    plt.title('Chromagram')
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()
    plt.close()
plot_chromagram(file_path, output_folder)
```

```python
def plot_spectral_rolloff(file_path, output_folder, sr=22050):
    y, sr = librosa.load(file_path, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    output_file = os.path.join(output_folder, "spectral_rolloff_plot.png")
    
    plt.figure(figsize=(15, 9))
    plt.plot(rolloff[0], label='Spectral Roll-off', color='purple')
    plt.title('Spectral Roll-off')
    plt.xlabel('Frames')
    plt.ylabel('Frequency (Hz)')
    plt.grid()
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()
    plt.close()
plot_spectral_rolloff(file_path, output_folder)
```

```python
def plot_spectral_centroid(file_path, output_folder, sr=22050):
    y, sr = librosa.load(file_path, sr=sr)
    centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    output_file = os.path.join(output_folder, "spectral_centroid_plot.png")
    plt.figure(figsize=(15, 9))
    plt.plot(centroid[0], label='Spectral Centroid', color='green')
    plt.title('Spectral Centroid')
    plt.xlabel('Frames')
    plt.ylabel('Frequency (Hz)')
    plt.tight_layout()
    plt.savefig(output_file)
    plt.show()
    plt.close()
plot_spectral_centroid(file_path, output_folder)
```
